```python
# Importing dataset
import pandas as pd

movies = pd.read_csv(filepath_or_buffer="data/movies.csv").loc[:, ["original_title", "overview"]]
movies.head(10)
```

```python
# Word Tokenizing the overviews of movies
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize
tokens = word_tokenize(" ".join(movies['overview']))
```

```python
# Stemming
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()
sentence = "Programmers program with programming languages"
tokens = word_tokenize(sentence)

stemmed = [stemmer.stem(word) for word in tokens]
stemmed  
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download WordNet and punkt if you haven't already
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# Initialize the lemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# Sample sentence
sentence = "He was running and eating at the same time. He has a bad habit of swimming after plaing long hours in the Sun."

# Tokenize the sentence into words
words = word_tokenize(sentence)

# Lemmatize each word in the sentence with 'v' as the POS tag for verbs
lemmatized_words = [wordnet_lemmatizer.lemmatize(word, pos='v') for word in words]

# Join the lemmatized words back into a sentence
lemmatized_sentence = ' '.join(lemmatized_words)

print(lemmatized_sentence)
```
