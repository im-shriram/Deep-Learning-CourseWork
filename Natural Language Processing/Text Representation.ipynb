```python
# Importing dataset
import pandas as pd

movies = pd.read_csv(filepath_or_buffer="data/movies.csv").loc[:, ["original_title", "overview"]]
movies.head(10)
```

```python
# One-Hot Encoding
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts(" ".join(movies["overview"]).split(" ")) # List of words
sequences = tokenizer.texts_to_sequences(" ".join(movies["overview"]))

tokenizer.word_index
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

def one_hot_encoding(text, max_len: int = 50):
    tokens = tokenizer.texts_to_sequences(text.split(" ")) # Expect list of words not complete sentence
    tokens = [(token[0] if len(token) else 0) for token in tokens]
    padded_tokens = pad_sequences([tokens], maxlen=max_len, padding='post', value=0)
    return to_categorical(padded_tokens, num_classes=len(tokenizer.word_index))[0]

movies.loc[:5, "overview"].apply(lambda overview: one_hot_encoding(overview, max_len=175))
```

---

```python
# Bag of Words
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(lowercase = True, stop_words = 'english', analyzer = 'word', ngram_range = (1, 1), binary = False)
"""
    unigram: (1, 1) - single word
    bigram: (1, 2) - single and combination of 2 adjecent words
    trigram: (1, 3) - single, combination of 2 adjecent words and combination of 3 adjecent words
"""
bow = cv.fit_transform(movies['overview'])

cv.vocabulary_
```

```python
# Total words
len(cv.vocabulary_)
```

```python
# Encoding first 5 overviews of movies
bow[:5].toarray()
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfimovies = TfidfVectorizer()
encoded = tfimovies.fit_transform(movies['overview'])
```

```python
print(encoded[5].toarray()) # When you print encoded[5].toarray(), NumPy truncates the display of large arrays. Since TF-IDF vectors are extremely sparse (most words in your dataset don't appear in one specific movie's overview), the beginning and end of the vector are likely zeros.

# Check the number of non-zero elements
print(f"Number of non-zero entries: {encoded[5].nnz}")
```

```python
print(tfimovies.idf_)
```

`idf_` (Inverse Document Frequency): This is a global weight for each word. It tells you how important/rare a word is across your entire dataset. It's the same for every movie.

`tf` (Term Frequency): This measures how many times a word appears in a specific movie overview.
tf-idf (The Result): This is tf * idf. It highlights words that are frequent in a specific document but rare across the whole dataset (providing the best representation).
