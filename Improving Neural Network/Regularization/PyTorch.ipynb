In PyTorch, the application of L1 and L2 regularization differs primarily in how they are integrated into the training process. While L2 is a built-in feature of most optimizers, L1 must typically be added manually to your loss calculation. [1, 2, 3, 4]  
1. Applying L2 Regularization (Weight Decay) 
The most efficient and recommended way to apply L2 regularization is through the  parameter in your optimizer. This automatically adds a penalty proportional to the sum of the squared weights to the loss. 

• Implementation: Use the  argument when initializing your optimizer. 
• Effect: It prevents coefficients from becoming excessively large, reducing model complexity. [5, 7, 9, 10, 11]  

2. Applying L1 Regularization (Lasso) 
PyTorch does not have a built-in  for L1 regularization in its standard optimizers. You must manually calculate the sum of absolute weights and add it to your total loss before backpropagation. 
```python
    l1_lambda = 0.001
    l1_norm = sum(p.abs().sum() for p in model.parameters())

    loss = criterion(outputs, targets) + l1_lambda * l1_norm
    loss.backward()
    optimizer.step()
```
• Implementation: 
• Effect: It encourages sparsity, often driving unimportant weights to exactly zero, which acts as a form of automatic feature selection. [14, 15, 16, 17, 18]  

3. Applying Both (Elastic Net) 
If you want to leverage both methods, you can combine them manually by adding both penalty terms to your loss. 
```python
    l1_lambda = 0.001
    l2_lambda = 0.001

    # Calculate norms
    l1_norm = sum(p.abs().sum() for p in model.parameters())
    l2_norm = sum(p.pow(2.0).sum() for p in model.parameters())

    # Total loss
    loss = criterion(outputs, targets) + (l1_lambda * l1_norm) + (l2_lambda * l2_norm)
```
