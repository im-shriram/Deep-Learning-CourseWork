#### Momentum Update Formulations

**1. The Alternative Implementation**
In this version, the learning rate is applied directly to the velocity during the parameter update step, rather than scaling the gradient before the momentum accumulation.

$$
\begin{aligned}
v_{t} &= \mu \cdot v_{t - 1} + g_{t} \\
p_{t+1} &= p_t - \text{lr} \cdot v_{t}
\end{aligned}
$$

**2. The Sutskever et al. Implementation**
This is the standard formulation found in "On the Importance of Initialization and Momentum in Deep Learning" and frameworks like PyTorch (when `momentum` is set without `nesterov=True` logic variations) or TensorFlow. Here, the learning rate scales the gradient immediately.

$$
\begin{aligned}
v_{t} &= \mu \cdot v_{t - 1} + \text{lr} \cdot g_{t} \\
p_{t+1} &= p_t - v_{t}
\end{aligned}
$$

*Note: While mathematically similar in behavior, the placement of `lr` changes the effective scale of the accumulated velocity vector `v`. In the first case, `v` represents the raw gradient accumulation, while in the second, `v` represents the accumulated update step.*
